```r
# Get the number of columns in the data frame
num_cols <- ncol(tab)

# Create an empty vector to store the number of missing values for each column
missing_values <- numeric(num_cols)

# Loop over the columns and calculate the number of missing values for each column
for (i in 1:num_cols) {
  missing_values[i] <- sum(is.na(tab[[i]]))
}

# Print the number of missing values for each column
print(missing_values)
```
